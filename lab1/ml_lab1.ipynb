# Q1. Generate a data set with numerical and categorical values. (For Example: Personal Details- name, age, qualification, city, salary, working  company). Use data frame, random methodsâ€¦
## a) For the salary, Find mean, median, mode. Find variance, standard deviation
## b) For the salary, check it skewed or symmetric
## c)Find correlation between age and salary.
## d)Count missing values per column.  Fill missing values using mean (numeric) and mode (categorical)

```python
import pandas as pd
import numpy as np
np.random.seed(42)
n = 20
names = ["Arjun", "Meera", "Kabir", "Sanya", "Nikhil", "Riya", "Aditya", "Kavya",
"Manav", "Tanya"]
cities = ["Pune", "Kolkata", "Jaipur", "Ahmedabad", "Chandigarh"]
qualifications = ["BCA", "MCA", "B.Com", "M.Com", "PhD"]
companies = ["Accenture", "IBM", "Deloitte", "Microsoft", "Flipkart"]
data = {
"Name": np.random.choice(names, n),
"Age": np.random.randint(21, 40, n),
"Qualification": np.random.choice(qualifications, n),
"City": np.random.choice(cities, n),
"Salary": np.random.randint(25000, 120000, n),
"Company": np.random.choice(companies, n)
}
df = pd.DataFrame(data)
df
```

```python
df.loc[np.random.choice(df.index, 3), "Salary"] = np.nan
df.loc[np.random.choice(df.index, 2), "Age"] = np.nan
df.loc[np.random.choice(df.index, 2), "City"] = np.nan
df
```

```python
salary_mean = df["Salary"].mean()
salary_median = df["Salary"].median()
salary_mode = df["Salary"].mode()[0]
salary_variance = df["Salary"].var()
salary_std = df["Salary"].std()
print("Mean Salary:", salary_mean)
print("Median Salary:", salary_median)
print("Mode Salary:", salary_mode)
print("Variance of Salary:", salary_variance)
print("Standard Deviation of Salary:", salary_std)
```

```python
salary_skewness = df["Salary"].skew()
print("Skewness of Salary:", salary_skewness)
if salary_skewness > 0:
  print("Right Skewed")
elif salary_skewness < 0:
  print("Left Skewed")
else:
  print("Symentric")
```

```python
correlation = df[["Age", "Salary"]].corr()
print("Correlation Matrix:")
print(correlation)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
df
```

```python
df["City"].fillna(df["City"].mode()[0], inplace=True)
df
```

# Q2. Create a dataframe â€˜daâ€™ for a dataset. Transform â€˜daâ€™ into another dataframe â€˜da1â€™ by replacing the entries of â€˜daâ€™ as follows.

## â€“ If the entries in the column are numerical, first 25% of the entries in each column are replaced by the mean of the values in that column, second 25% of the entries in each column are replaced by the ((maximum value of that column), third 25% of the entries in each column are replaced by the (minimum value of that column) and the last, 25% of the entries in each column are replaced by the ((mean value of that column)/2)
## â€“ If the entries in the column are non-numerical, last 50% of the entries in each are replaced by the entry that occurs least in that column, first  50% of the entries in each are replaced by the entry that has highest frequency in that column.
## â€“ If the entries in a column has both numerical or non-numerical entries, replace all the non-numerical entries with zeros

```python
import pandas as pd
import numpy as np
np.random.seed(1)
da = pd.DataFrame({
"Age": np.random.randint(20, 40, 20),
"Salary": np.random.randint(30000, 90000, 20),
"City": np.random.choice(["Delhi", "Mumbai", "Pune"], 20),
"Department": np.random.choice(["HR", "IT", "Sales"], 20),
"Mixed_Column": np.random.choice([10, 20, "A", "B"], 20)
})
da
```

```python
da1 = da.copy()
n = len(da)
q1 = int(0.25 * n)
q2 = int(0.50 * n)
q3 = int(0.75 * n)
for col in da.columns:
  if pd.api.types.is_numeric_dtype(da[col]):
    da1[col] = da1[col].astype(float)
    mean_val = da[col].mean()
    max_val = da[col].max()
    min_val = da[col].min()
    da1.iloc[:q1, da.columns.get_loc(col)] = mean_val
    da1.iloc[q1:q2, da.columns.get_loc(col)] = max_val
    da1.iloc[q2:q3, da.columns.get_loc(col)] = min_val
    da1.iloc[q3:, da.columns.get_loc(col)] = mean_val / 2
  elif da[col].apply(lambda x: isinstance(x, str)).all():
    most_frequent = da[col].mode()[0]
    least_frequent = da[col].value_counts().idxmin()
    da1.iloc[:q2, da.columns.get_loc(col)] = most_frequent
    da1.iloc[q2:, da.columns.get_loc(col)] = least_frequent
  else:
    da1[col] = da[col].apply(lambda x: x if isinstance(x, (int, float)) else 0)
print("Original DataFrame (da):\n")
print(da)
print("\nTransformed DataFrame (da1):\n")
print(da1)
```
