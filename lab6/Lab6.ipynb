```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder,KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("adult.csv")
df
```

```python
features=['age','education.num','marital.status','capital.gain','capital.loss','hours.per.week']
X=df[features]
y=df['income']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1172)
```

```python
num_cols=['age','capital.gain','capital.loss','hours.per.week']
cat_cols=['marital.status']
passthrough_cols=['education.num']

preprocessor =ColumnTransformer(
    transformers=[
        ('num',KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile'),num_cols),
        ('cat',OrdinalEncoder(),cat_cols),
        ('pass','passthrough',passthrough_cols)
    ]
)
```

```python
X_train_processed=preprocessor.fit_transform(X_train)
```

```python
X_test_processed = preprocessor.transform(X_test)
```

```python
cart_model = DecisionTreeClassifier(criterion='gini', random_state=1172)
cart_model.fit(X_train_processed, y_train)
```

```python
id3_model = DecisionTreeClassifier(criterion='entropy', random_state=1172)
id3_model.fit(X_train_processed, y_train)
```

```python
nb_model = CategoricalNB()
nb_model.fit(X_train_processed, y_train)
```

```python
y_pred_cart = cart_model.predict(X_test_processed)
accuracy_score(y_test, y_pred_cart)
```

```python
y_pred_id3 = id3_model.predict(X_test_processed)
accuracy_score(y_test, y_pred_id3)
```

```python
y_pred_naive = nb_model.predict(X_test_processed)
accuracy_score(y_test, y_pred_naive)
```
